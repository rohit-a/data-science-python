# Machine Learning Overview

## Supervised Learning

- Model traiing on labelled data



#### Classification

- Predicting discrete values
- Identifying input data as a member of particular class
- Training model on dataset of pre-labeled animal images (Cats and Dogs). The algorithm is then evaluated by how accurately it can correctly classify new images
- eg: k-Nearest Neighbors, Logistic Regression

#### Regression
- Predicting Continuous values
- Predict 'y' value given list of 'x' values
- Predicting cost of apartment based on location, number of bedrooms, city etc 
- eg: Linear Regression

## Unsupervised Learning



- Model trained on unlabelled data.
- Pattern discovery unguided by labels
- Model automatically finds structure in the data by extracting useful features and analyzing its structure.
- Customer Segmentation, Recommendation Systems
- eg: Clustering, Dimensionality Reduction

## Semi Supervised Learning

- Mixture of labelled and unlabelled data
- There is a desired prediction problem but the model must learn the structures to organize the data as well as make predictions
- eg: predicting disease based on CT Scans, where an expert labels few CT Scans


## Reinforcement Learning



- AI agents are attempting to find the optimal way to accomplish a particular goal, or improve performance on a specific task. 
- Each action that goes toward the goal, it receives a reward. 
- Overall aim: predict the best next step to take to earn the biggest final reward.
- Agent relies both on learnings from past feedback and exploration of new tactics that may present a larger payoff. This involves a long-term strategy — just as the best immediate move in a chess game may not help you win in the long run, the agent tries to maximize the cumulative reward.
- Iterative approach
- eg: Game AI, Robot Navigation
