## Typical Process for Implementing ML

1. Asking the Buisness Question (Most Important Step) 
2. Define Success Criteria, Supervised vs Unsupervised Learning
3. Find/Collect Data Sources to help solve the problem
4. Import Data : Pandas
5. Exploratory Data Analysis <-> Cleaning Data : Pandas, matplotlib, seaborn
6. Feature Extraction (eg: process text/image data) : sklearn.feature_extraction
7. Feature Selection (eg: NMF, Dimenssionality Reduction) : sklearn.feature_selection
8. Holdout data for model evaluation after traing the model (10%-20% data) : sklearn.model_selection.train_test_split
9. Apply ML algorithms: Classification, Regression, Clustering
10. Evaluate model performance - Scoring, Hyperparamter tuning
11. Test model on Hold out data
12. Repeat 7-11 for best success criteria
13. Persist model
14. Deployment
15. Periodic Evaluation and updates

### Importing Data

```python
# Importing datasets
import pandas as pd 
import numpy as np

df_credits=pd.read_csv('tmdb_5000_credits.csv')
df_movies=pd.read_csv('tmdb_5000_movies.csv')
print(df_credits.shape)
df_credits.head()
```

```python
print(df_movies.shape)
df_movies.head()
```

```python
# Merging both datasets
df_movies= df_movies.merge(df_credits.rename(columns={"movie_id":"id"}), on="id")
df_movies.head()
```

```python
#Inspecting the data frame
df_movies.info()
df_movies.isnull().sum()
```

### Recommending movies based on cast, crew, keyword and genres

We are going to build a recommender based on the following metadata: the 3 top actors, the director, related genres and the movie plot keywords.

From the cast, crew and keywords features, we need to extract the three most important actors, the director and the keywords associated with that movie. Right now, our data is present in the form of "stringified" lists , we need to convert it into a safe and usable structure.

```python
# Look at feature columns
features = ['cast', 'crew', 'keywords', 'genres']
df_movies[features]
```

```python
# Parse the stringified features into their corresponding python objects
from ast import literal_eval
for feature in features:
    df_movies[feature] = df_movies[feature].apply(literal_eval)
df_movies[features]
```

```python
# Exract director's name
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan

# Get top 3 cast, keywords and genres
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        #Check if more than 3 elements exist. If yes, return only first three. If no, return entire list.
        if len(names) > 3:
            names = names[:3]
        return names

    #Return empty list in case of missing/malformed data
    return []

df_movies['director'] = df_movies['crew'].apply(get_director)
features_top3 = ['cast', 'keywords', 'genres']
for feature in features_top3:
    df_movies[feature] = df_movies[feature].apply(get_list)
```

```python
# Look at sample rows
df_movies[['original_title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

Clean up data and convert to lower case

```python
# Function to convert all strings to lower case and strip names of spaces
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        #Check if director exists. If not, return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''

# Apply clean_data function to your features.
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    df_movies[feature] = df_movies[feature].apply(clean_data)
df_movies[['original_title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

Create metadata column that has all text ('cast', 'keywords', 'director', 'genres')

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df_movies['soup'] = df_movies.apply(create_soup, axis=1)
df_movies['soup']
```

Using CountVectorizer to convert text to matrix of token counts
(https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html)

```python
# Import CountVectorizer and create the count matrix
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df_movies['soup'])
count_matrix
```

```python
count.get_feature_names()
```

```python
print(count_matrix.shape)
count_matrix.toarray()
```

Using cosine similarity to to get similar movies
(https://en.wikipedia.org/wiki/Cosine_similarity)
(https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise.cosine_similarity.html)

```python
# Compute the Cosine Similarity matrix based on the count_matrix
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(count_matrix, count_matrix)
print(cosine_sim.shape)
cosine_sim
```

```python
#Construct a reverse map of indices and movie titles
indices = pd.Series(df_movies.index, index=df_movies['original_title']).drop_duplicates()
indices
```

```python
# Function that takes in movie title as input and outputs most similar movies
def get_recommendations(title, cosine_sim=cosine_sim):
    # Get the index of the movie that matches the title
    idx = indices[title]

    # Get the pairwsie similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return df_movies['original_title'].iloc[movie_indices]
```

```python
get_recommendations('The Dark Knight Rises', cosine_sim)
```

```python
get_recommendations('Tangled', cosine_sim)
```

```python
get_recommendations('Harry Potter and the Half-Blood Prince', cosine_sim)
```
